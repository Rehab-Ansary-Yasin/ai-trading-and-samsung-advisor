```python
import sqlite3

conn = sqlite3.connect("samsung.db")
cur = conn.cursor()

cur.execute("""
CREATE TABLE IF NOT EXISTS phones (
    model TEXT,
    display TEXT,
    battery TEXT,
    camera TEXT,
    price INTEGER
)
""")

phones = [
    ("Samsung Galaxy S23 Ultra", "6.8 AMOLED", "5000mAh", "200MP", 1199),
    ("Samsung Galaxy S22 Ultra", "6.8 AMOLED", "5000mAh", "108MP", 999),
    ("Samsung Galaxy S21 FE", "6.4 AMOLED", "4500mAh", "12MP", 699)
]

cur.executemany("INSERT INTO phones VALUES (?,?,?,?,?)", phones)
conn.commit()
conn.close()
```

```python
from fastapi import FastAPI

app = FastAPI()

@app.post("/ask")
def ask(q: dict):
    return {
        "answer": "Samsung Galaxy S23 Ultra has better camera and battery life than S22 Ultra."
    }
```

```python
import threading
import uvicorn

def run_server():
    uvicorn.run(app, host="127.0.0.1", port=8000)

thread = threading.Thread(target=run_server, daemon=True)
thread.start()
```
